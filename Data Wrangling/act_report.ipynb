## WeRateDogs act_report

### Analysis and Viisualization
After Gathering, Assessing and Cleaning datasets about a Twitter account 'WeRateDogs', I decided to draw more insight on the project by visualizing my insights for clearer understanding. 


#### 1.	Insight on the most popular dog name

The bar chart indicates that 'Charlie' is the most popular dog name used, followed by Lucy.

#### 2.	Insight on the major twitter source

The sources used to engage the WeRateDogs are Iphone, Web and Deck. The bar chart indidcates that Twitter for Iphone is the most used engagement source with over 1400 tweets while Twitter for Deck and Web did not get up to 200 tweets.

#### 3.	Insight on the dog stages distribution


There are four dog stages in WeRateDogs twitter page; pupper, doggo, puppo and floofer. According to the pie chart, Pupper is the most popular dog stage with the most tweets. Puppers are actually very cute as that is the earliest stage of a dog's life.

#### 4.	Insight on whether the favorite_count has an effect on retweet_count i.e a correlation

There is a strong positive correlation of 0.91084203308148481 i.e 91% between the likes and retweets. It means that there is a 91% chance of retweeting and engaging more on the tweet of a dog after liking the post.

### CONCLUSION
   1. Charlie is the most popular dog name, followed by lucy.
   2. Twitter for Iphone is the most used by users.
   3. Pupper was the most popular dog stage, not surprising as that is the earliest dog stage.
   4. There is a 91% positive correlation between the likes and retweets, It is obvious that as users like the image, they tend to retweet the image.
